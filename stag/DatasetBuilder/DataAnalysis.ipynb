```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
# CSV 파일 경로
csv_file_path = './BTC_BackTest.csv'

# CSV 파일을 읽어와서 데이터프레임 생성
df = pd.read_csv(csv_file_path)

# describe() 함수를 사용하여 데이터 요약 출력
description = df['POSITION'].value_counts()

# 출력
print(description,df['risk'].describe())
print(description[0]/description[2])
column_name = '20 배 선물 투자시 비트코인 투자에서 10%를 벌 수 있는 통계'
font_path = "C:\\Windows\\Fonts\\malgun.ttf"  # 한글 폰트 파일 경로 (Windows 기준, 다른 운영체제는 경로가 다를 수 있음)
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.bar(['20%이상 벌지 못한다','롱','숏'], description.values)

# 그래프 제목과 축 레이블 설정
plt.title(column_name)
plt.xlabel(column_name)
plt.ylabel('Count')

# 그래프 표시
plt.show()
```
